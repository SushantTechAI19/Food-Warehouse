##### Food Warehouse Temperature Monitoring Analysis #####



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating sample temperature data
np.random.seed(60)
days = pd.date_range(start="2025-06-09", periods=110)
temperature = np.random.normal(loc=20, scale=4,
size=110) # Ideal cold storage temp ≈ 8°C
data = pd.DataFrame({
"Date": days, "Temperature (°C)": temperature
})
print(data.head())
```

```python
# Exploratory Data Analysis
print("\nDataset Information")
print(data.info())
print("\nDescriptive Statistics")
print(data.describe())
```

```python
#Temperature Trend Analysis (line plot)
plt.figure(figsize=(15, 6))
plt.plot(data["Date"], data["Temperature (°C)"], marker='o')

plt.axhline(y=10, color='r', linestyle='--',
            label='Upper Safe Limit (12°C)')

plt.axhline(y=4, color='g', linestyle='--',
            label='Lower Safe Limit (2°C)')

plt.title("Food Warehouse Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Spoilage Prevention Analysis
def spoilage_risk(temp):
  if temp <2 or temp>12:
    return "Risk"
  else:
    return "Safe"
    data["Spoilage Risk"] = data["Temperature (°C)"].apply(spoilage_risk)
    print(data["Spoilage Risk"]=="Risk")
```

```python
#Variance Analysis
variance = data["Temperature (°C)"].var()
std_dev = data["Temperature (°C)"].std()
print(f"Temperature Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
#Daily Temperature Change Analysis
data["Daily Change"] = data["Temperature (°C)"].diff()
plt.figure(figsize=(15, 5))
plt.plot(data["Date"], data["Daily Change"],
color='purple', marker='o')
plt.title("Daily Temperature Change")
plt.xlabel("Date")
plt.ylabel("Change in Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
if variance > 5:
 print("A Warning: High temperature variance detected. Risk of spoilage.")
else:
 print(" Temperature conditions are stable.")
```
